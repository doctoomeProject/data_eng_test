```python
import pandas as pd

CIS_COMPO_bdpm = pd.read_csv("../../drugs_project/data/CIS_COMPO_bdpm.txt", sep="\t", header=None, dtype=str, encoding="latin1")
CIS_COMPO_bdpm

composition_columns = [
    "cis_code",                          # Links to the specialty (foreign key)
    "pharma_element",               # Pharmaceutical element designation
    "substance_code",                     # Substance code
    "substance_name",             # Substance name
    "dosage",                   # Dosage amount
    "dosage_ref",                   # Dosage reference (e.g. “per tablet”)
    "component_type",                   # “SA” (active substance) or “ST” (therapeutic fraction)
    "link_num"                  # Linking number between substances/fractions
]
CIS_COMPO_bdpm.columns=composition_columns
```

```python
CIS_COMPO_bdpm['cis_code'] = CIS_COMPO_bdpm['cis_code'].astype(int)
CIS_COMPO_bdpm['substance_code'] = CIS_COMPO_bdpm['substance_code'].astype(int)
CIS_COMPO_bdpm['link_num'] = CIS_COMPO_bdpm['link_num'].astype(int)
```

```python
import mysql.connector

# Connect to the MySQL container
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="user",
    password="pass",
    database="source_database"
)
cur = conn.cursor()

# Prepare an INSERT statement
insert_sql = """
INSERT INTO CIS_COMPO_bdpm (
    cis_code, pharma_element, substance_code, substance_name,
    dosage, dosage_ref, component_type, link_num
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
"""

# Insert each row
for _, row in CIS_COMPO_bdpm.iterrows():
    cur.execute(insert_sql, tuple(row))

conn.commit()
cur.close()
conn.close()
print("✅ Data successfully inserted into MySQL")
```
